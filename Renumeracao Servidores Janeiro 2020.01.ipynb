# Salarios dos Servidores Brasileiros - Janeiro 2020

```python
df=pd.read_csv('202001_Remuneracao.csv',sep=';',encoding='ISO-8859-1')
```

```python
#Mudando a configuração de "," para "."
df.iloc[:,5:]=df.iloc[:,5:].apply(lambda x: x.str.replace(',','.'))
```

```python
#Mudando o dtype para float
df.iloc[:,5:]=df.iloc[:,5:].astype(float)
```

```python
#Algumas informações relevantes sobre o dataset - já podemos ver algumas informações interessantes..como podemos ter remumeração bruta após deduções negativa?
df.describe()
```

```python
#Vou deletar os salarios negativos para não afetar muito a análise
df=df[df['REMUNERAÇÃO APÓS DEDUÇÕES OBRIGATÓRIAS (R$)']>0]
```

```python
#Começando a análise - qual a média/mediana salarial após deduções dos servidores?
df['REMUNERAÇÃO APÓS DEDUÇÕES OBRIGATÓRIAS (R$)'].mean()
```

A média salarial é bem alta se compararmos com o salário mínimo e também maior que o salário de engenheiro em muitos casos...

```python
# Qual o top 10 salários após deduções?
df.sort_values('REMUNERAÇÃO APÓS DEDUÇÕES OBRIGATÓRIAS (R$)',ascending=False)[['NOME','REMUNERAÇÃO APÓS DEDUÇÕES OBRIGATÓRIAS (R$)']].head(10)
```

 Parece que somente no top 10 salário, TODOS ganharam mais que o teto salarial constitucional de 39 mil R$.

```python
#Quantas pessoas ganharam mais que o teto?
df[df['REMUNERAÇÃO APÓS DEDUÇÕES OBRIGATÓRIAS (R$)']>39000][['NOME','REMUNERAÇÃO APÓS DEDUÇÕES OBRIGATÓRIAS (R$)']].sort_values('REMUNERAÇÃO APÓS DEDUÇÕES OBRIGATÓRIAS (R$)',ascending=False)
```

2742 servidores tiveram uma remuneração após deduções maior que o teto constitucional de 39.000 R$....

```python
#Vamos dar uma olhada nas verbas indezatórias?
df['TOTAL DE VERBAS INDENIZATÓRIAS (R$)(*)'].sum()/1000000
```

 Só em verbas indenizatórias o governo gastou 365 milhões de reais..

```python
#Vamos dar uma olhada no top 20 servidores que mais ganharam verbas?
df.sort_values('TOTAL DE VERBAS INDENIZATÓRIAS (R$)(*)',ascending=False)[['NOME','TOTAL DE VERBAS INDENIZATÓRIAS (R$)(*)']].head(20)
```
